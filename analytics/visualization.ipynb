```python
tips = px.data.tips()
tips
tips.columns.tolist()
```

```python
# pie chart
gc= tips['sex'].value_counts()
px.pie(gc, gc.index, 
       gc.values,
       title= 'gender distribution in data',
        template='plotly_dark',
          )
```

```python
# donut chart
gc= tips['sex'].value_counts()
px.pie(gc, gc.index, 
       gc.values,
       title= 'gender distribution in data',
        template='plotly_dark',
        hole=0.5
          )
```

```python
def get_pie_chart(df, column,hole = 0.5,template='plotly_dark'):
   count= df[column].value_counts()
   fig =px.pie(count, count.index, 
       count.values,
       title=f'{column} distribution in data',
       template= template,
       hole=hole)
   return fig
```

```python
get_pie_chart(tips, 'sex')
```

```python
get_pie_chart(tips,'smoker')
```

```python
get_pie_chart(tips,'day')
```

```python
get_pie_chart(tips, 'time')
```

```python
get_pie_chart(tips, 'size')
```
